# **Creating the train set**

```python
import pandas as pd
train_pd = pd.read_csv('train.csv')
train_pd
```

```python
#we will use Survived, Pclass, Sex, Age, SibSp, Parch, Fare, and Embarked
train_pd= train_pd[['Survived', 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']]
train_pd
```

```python
#to remove NaN data
train_pd = train_pd.dropna()
train_pd
```

```python
train_pd.describe()
```

# **Data Conversion**
**Pclass:**
*   1-> [1,0,0]
*   2-> [0,1,0]
*   3-> [0,0,1]

**Sex:**
*   male-> 0
*   female-> 1





```python
import numpy as np
from tensorflow.keras.utils import to_categorical
#convert pclass into the categorical form
train_pclass = train_pd['Pclass'].to_numpy()
as_categorical_train_pclass = to_categorical(train_pclass - 1, num_classes=3)
as_categorical_train_pclass
```

```python
#to convert the sex from string to boolean
def sex_str_to_binary(s):
  if s == 'male':
    return 0
  else:
    return 1
```

```python
as_binary_train_sex = train_pd['Sex'].apply(sex_str_to_binary).to_numpy()
as_binary_train_sex[:10]
```

```python
#convert age to numpy array
train_age_np = train_pd['Age'].to_numpy()
#scale the Age values
train_age =(train_age_np-train_pd['Age'].mean()) / train_pd['Age'].std()
train_age[:10]
```

```python
#convert SibSp to numpy array
train_sibsp_np = train_pd['SibSp'].to_numpy()
#scale the SibSp values
train_sibsp =(train_sibsp_np-train_pd['SibSp'].mean()) / train_pd['SibSp'].std()
train_sibsp[:10]
```

```python
#convert Parch to numpy array
train_parch_np = train_pd['Parch'].to_numpy()
#scale the Parch values
train_parch =(train_parch_np-train_pd['Parch'].mean()) / train_pd['Parch'].std()
train_parch[:10]
```

```python
#convert fare to numpy array
train_fare_np = train_pd['Fare'].to_numpy()
#scale the Fare values
train_fare =(train_fare_np-train_pd['Fare'].mean()) / train_pd['Fare'].std()
train_fare[:10]
```

```python
Y_train = train_pd['Survived'].to_numpy()
Y_train[:10]
```

# **Creating the model**

we will ues scikit learns linear model logistic regression to converts all inputs to continuous values and after that we convert the predicting result 0 or 1 to a probability.

```python
from sklearn.linear_model import LogisticRegression
```

```python
as_categorical_train_pclass.shape,as_binary_train_sex.reshape(-1, 1).shape
```

```python
train_age.shape
```

```python
train_fare.shape
```

```python
train_parch.shape
```

```python
train_sibsp.shape
```

```python
#that is our training dataset
X_train=np.concatenate((as_categorical_train_pclass,
                as_binary_train_sex.reshape(-1, 1),
                train_age.reshape(-1,1),
                train_sibsp.reshape(-1,1),
                train_parch.reshape(-1,1),
                train_fare.reshape(-1,1)),axis=1)
X_train
```

```python
model = LogisticRegression(random_state=0).fit(X_train, Y_train)
model.score(X_train, Y_train)
```

# **Creating the test set**

```python
test_pd = pd.read_csv('test.csv')
test_pd
```

```python
test_pd= test_pd[['PassengerId', 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare']]
test_pd
```

```python
test_pclass = test_pd['Pclass'].to_numpy()
as_categorical_test_pclass = to_categorical(test_pclass - 1, num_classes=3)
as_categorical_test_pclass[:10]
```

```python
as_binary_test_sex = test_pd['Sex'].apply(sex_str_to_binary).to_numpy()
as_binary_test_sex[:10]
```

```python
#convert age to numpy array
test_age_np = test_pd['Age'].to_numpy()
#scale the Age values
test_age =(test_age_np-test_pd['Age'].mean()) / test_pd['Age'].std()
test_age[:10]
```

```python
#convert SibSp to numpy array
test_sibsp_np = test_pd['SibSp'].to_numpy()
#scale the SibSp values
test_sibsp =(test_sibsp_np-test_pd['SibSp'].mean()) / test_pd['SibSp'].std()
test_sibsp[:10]
```

```python
#convert Parch to numpy array
test_parch_np = test_pd['Parch'].to_numpy()
#scale the Parch values
test_parch =(test_parch_np-test_pd['Parch'].mean()) / test_pd['Parch'].std()
test_parch[:10]
```

```python
#convert fare to numpy array
test_fare_np = test_pd['Fare'].to_numpy()
#scale the Fare values
test_fare =(test_fare_np-test_pd['Fare'].mean()) / test_pd['Fare'].std()
test_fare[:10]
```

```python
X_test=np.concatenate((as_categorical_test_pclass,
                as_binary_test_sex.reshape(-1, 1),
                test_age.reshape(-1,1),
                test_sibsp.reshape(-1,1),
                test_parch.reshape(-1,1),
                test_fare.reshape(-1,1)),axis=1)
X_test
```

# **Model Predicition**

```python
predictions=model.predict(X_test)
PassengerId=test_pd['PassengerId'].to_numpy()
sub_df=pd.DataFrame({'PassengerId':PassengerId,
                     'Survived':predictions})
sub_df
```

```python
sub_df.to_csv('submission.csv',index=False)
```
